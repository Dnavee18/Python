# Web_Scraping

```python
import requests # to get/fetch data from the target website
import pandas as pd # for data manipulation
from bs4 import BeautifulSoup # to filter the response based on pattern
```

```python
requests.get("https://www.dominos.co.in/store-location/chennai")
```

```python
response = requests.get("https://www.dominos.co.in/store-location/chennai")
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
# <p class="city-main-sub-title">
```

```python
location = soup.find_all("p",{"class":"city-main-sub-title"})
location = [i.text for i in location]
location
```

```python
# <P class ="grey-text mb-0">
```

<p class="grey-text mb-0">

```python
address  = soup.find_all("p",{"class":"grey-text mb-0"})
for i in address:
  print(i.text)
```

```python
address =  [i.text for i in address]
address
```

```python
timing = soup.find_all("div", {"class":"col-xs-9 col-md-9 pl0 search-grid-right-text"})
timing
```

```python
timing = soup.find_all("div", {"class":"col-xs-9 col-md-9 pl0 search-grid-right-text"})
timing = [(i.text).strip("\n\tOpen Now")	for i in timing]
timing
```

```python
len(address)
```

```python
len(timing)
```

```python
len(location)
```

```python
data = {"Location":location,"Address" : address,"Service_timing":timing}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv("dominos.csv")
```
